```python
import numpy as np
import pandas as pd
```

```python
# 1: Nhập dữ liệu vào một danh sách
matrix = [
    ["D1", "Sunny", "High", "Weak", "No"],
    ["D2", "Sunny", "High", "Strong", "No"],
    ["D3", "Overcast", "High", "Weak", "Yes"],
    ["D4", "Rain", "High", "Weak", "Yes"],
    ["D5", "Rain", "Normal", "Weak", "Yes"],
    ["D6", "Rain", "Normal", "Strong", "No"],
    ["D7", "Overcast", "Normal", "Strong", "Yes"],
    ["D8", "Sunny", "High", "Weak", "No"],
    ["D9", "Sunny", "Normal", "Weak", "Yes"],
    ["D10", "Rain", "Normal", "Weak", "Yes"],
    ["D11", "Sunny", "Normal", "Strong", "Yes"],
    ["D12", "Overcast", "High", "Strong", "Yes"],
    ["D13", "Overcast", "Normal", "Weak", "Yes"],
    ["D14", "Rain", "High", "Strong", "No"]
]
data = pd.DataFrame(matrix, columns = ["Day","Outlook", "Humidity", "Wind", "PlayTennis"])
data
```

```python
c0 = data[['Day']]
c1 = data[['Outlook']]
c2 = data[['Humidity']]
c3 = data[['Wind']]
c4 = data[['PlayTennis']]
```

```python
# Hàm để chuyển giá trị thành số
def convert_to_numeric(value):
  # Nếu giá trị là 'High' hoặc 'Yes', trả về 1
  if value == 'High' or value == 'Yes':
    return 1
  # Nếu giá trị là 'Normal' hoặc 'No', trả về 0
  elif value == 'Normal' or value == 'No':
    return 0
  # Nếu giá trị là 'Sunny', trả về 2
  elif value == 'Sunny':
    return 2
  # Nếu giá trị là 'Overcast', trả về 3
  elif value == 'Overcast':
    return 3
  # Nếu giá trị là 'Rain', trả về 4
  elif value == 'Rain':
    return 4
  # Nếu giá trị là 'Weak', trả về 5
  elif value == 'Weak':
    return 5
  # Nếu giá trị là 'Strong', trả về 6
  elif value == 'Strong':
    return 6
  # Nếu không, trả về giá trị ban đầu
  else:
    return value

def mean(column):
      # Tạo một biến để lưu tổng
  total = 0
  # Tạo một biến để lưu số lượng
  count = 0
  # Duyệt qua mỗi giá trị trong cột
  for value in column:
    # Chuyển giá trị thành số
    value = convert_to_numeric(value)
    # Cộng giá trị vào tổng
    total += value
    # Tăng số lượng lên 1
    count += 1
  # Trả về tổng chia cho số lượng
  return total / count
```

```python
print(mean(np.array(c1)))
print(mean(np.array(c2)))
print(mean(np.array(c3)))
print(mean(np.array(c4)))
```

```python
# Hàm để tính độ lệch chuẩn của một cột số
def std(column):
  # Tạo một biến để lưu tổng bình phương sai số
  sum_of_squares = 0
  # Tạo một biến để lưu số lượng
  count = 0
  # Tính trung bình của cột
  mean_value = mean(column)
  # Duyệt qua mỗi giá trị trong cột
  for value in column:
    # Chuyển giá trị thành số
    value = convert_to_numeric(value)
    # Tính sai số giữa giá trị và trung bình
    error = value - mean_value
    # Bình phương sai số và cộng vào tổng
    sum_of_squares += error ** 2
    # Tăng số lượng lên 1
    count += 1
  # Trả về căn bậc hai của tổng chia cho số lượng
  return (sum_of_squares / count) ** 0.5
```

```python
print(std(np.array(c1)))
print(std(np.array(c2)))
print(std(np.array(c3)))
print(std(np.array(c4)))
```

```python
# Hàm để tính xác suất của mỗi giá trị duy nhất trong một cột
def prob(column):
  # Tạo một từ điển rỗng để lưu xác suất
  prob_dict = {}
  # Tạo một biến để lưu số lượng
  count = 0
  # Duyệt qua mỗi giá trị trong cột
  for value in column:
    # Tăng số lượng lên 1
    count += 1
    # Nếu giá trị đã có trong từ điển, tăng số lần xuất hiện lên 1
    if value in prob_dict:
      prob_dict[value] += 1
    # Nếu không, gán số lần xuất hiện bằng 1
    else:
      prob_dict[value] = 1
  # Duyệt qua mỗi khóa trong từ điển
  for key in prob_dict:
    # Chia số lần xuất hiện cho số lượng để tính xác suất
    prob_dict[key] = prob_dict[key] / count
  # Trả về từ điển xác suất
  return prob_dict
```

```python
print(prob(c1.values.flatten()))
print(prob(c2.values.flatten()))
print(prob(c3.values.flatten()))
print(prob(c4.values.flatten()))
```

```python
import math

# Hàm tính mean
def mean(numbers):
    return sum(numbers) / float(len(numbers))

# Hàm tính standard deviation
def stdev(numbers):
    avg = mean(numbers)
    variance = sum([pow(x - avg, 2) for x in numbers]) / float(len(numbers) - 1)
    return math.sqrt(variance)

# Hàm phân loại dữ liệu
def classify(data, column_index):
    column_data = [row[column_index] for row in data]
    column_data = [float(convert_to_numeric(i)) for i in column_data]
    
    mu = mean(column_data)
    sigma = stdev(column_data)
    
    classifications = []
    
    for value in column_data:
        # Áp dụng công thức Gaussian Naive Bayes
        probability_density_function_value = (1 / (math.sqrt(2 * math.pi * pow(sigma, 2)))) * \
                                             math.exp(-pow(value - mu, 2) / (2 * pow(sigma, 2)))
        
        # Định nghĩa một ngưỡng để phân loại
        if probability_density_function_value > 0:
            classifications.append('Yes') 
        else:
            classifications.append('No')
            
    return classifications
```
